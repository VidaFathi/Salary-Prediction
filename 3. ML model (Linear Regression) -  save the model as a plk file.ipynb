```python
import numpy as np
import sklearn
```

```python
#we use train_test_
from sklearn.model_selection import train_test_split
```

```python
# we should separate our dependant and indepandant variables:
#x= dependant variable
# we will drop the dependant variable which is our average salary  and put all of other variables in to 'x'
x= df2.drop(columns = 'Average_salary')
```

```python
# and put averge salary which is our dependant variable into 'y'
y = df2['Average_salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# now we want to devide our dataset into 2 categories, Training set and testing set.
# we do this with scikitlearn library
# to get the same random result we set the random_state 101
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
# we want to use linear regression:
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#we want to trainthe upper expression with Fit method:
lr = lr.fit(x_train, y_train)
```

```python
print(lr.coef_)
```

```python
# now we want to test the model. we give the x_test to the model and compare the result to y
#y_test
prediction = lr.predict(x_test)
```

```python
prediction
```

```python
y_test
```

```python
#now we're going to substract the prediction values by model from y_test to see whether or model is good or not:
residual = prediction - y_test
```

```python
residual
```

```python
import seaborn as sns
```

```python
sns.displot(residual, bins=50)
# we can see that the difference between the predictiona nd real data were around 0, then our model is great.
```

```python
#Mean Absolout Error(MAE) is a matrix that we can see how accurate is our predicted model.
```

```python
from sklearn import metrics
```

```python
print('MAE', metrics.mean_absolute_error(y_test, prediction))
```
